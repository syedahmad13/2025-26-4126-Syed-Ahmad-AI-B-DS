```python
t = (5, 10, 15, 20)

max_val = max(t)
min_val = min(t)

print("Max:", max_val)
print("Min:", min_val)
```

```python
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print("Union:", a | b)
print("Common:", a & b)
```

```python
marks = {"Math": 80, "CS": 90, "English": 70}

total = 0
for subject in marks:
    total += marks[subject]

print("Total:", total)
print("Average:", total / len(marks))
```
