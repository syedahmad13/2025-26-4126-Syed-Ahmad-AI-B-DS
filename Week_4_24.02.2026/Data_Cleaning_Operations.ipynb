```python
import pandas as pd
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
print("=== Original Dataset Info ===")
print(df.info())
print("\nMissing Values Before Cleaning:")
print(df.isnull().sum())
print("\nFirst 5 rows:\n", df.head())


# 1. Handle Missing Values
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

print("\n=== Missing Values After Filling ===")
print(df.isnull().sum())

# 2. Drop Cabin column (too many missing values)
df = df.drop(columns=['Cabin'])
print("\n=== Columns After Dropping 'Cabin' ===")
print(df.columns)

# 3. Remove Duplicate Rows
before_duplicates = df.shape[0]
df = df.drop_duplicates()
after_duplicates = df.shape[0]
print(f"\nRemoved {before_duplicates - after_duplicates} duplicate rows")

# 4. Convert Data Types
df['Survived'] = df['Survived'].astype('category')
df['Pclass'] = df['Pclass'].astype('category')
print("\n=== Data Types After Conversion ===")
print(df.dtypes)

# 5. Rename Columns to lowercase
df.columns = df.columns.str.lower()
print("\n=== Columns After Renaming ===")
print(df.columns)

# 6. Remove Outliers (Age using IQR method)
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

before_outliers = df.shape[0]
df = df[(df['age'] >= lower_bound) & (df['age'] <= upper_bound)]
after_outliers = df.shape[0]
print(f"\nRemoved {before_outliers - after_outliers} outlier rows based on Age")

# 7. Save Cleaned Dataset
df.to_csv("cleaned_titanic.csv", index=False)
print("\nData Cleaning Completed Successfully!")
print("Cleaned file saved as: cleaned_titanic.csv")

# Optional: View first 5 cleaned rows
print("\n=== First 5 rows of Cleaned Dataset ===")
print(df.head())
```
